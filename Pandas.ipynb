```python
import pandas as pd
mydataset = {
  'cars': ["BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2]
}
myvar=pd.DataFrame(mydataset)
print(myvar)
```

```python
#seires in pandas
a=[1,7,2]
myvar=pd.Series(a)
print(myvar)
```

```python
#labels
print(myvar[1])
```

```python
#create labels
myvar=pd.Series(a,index=["x","y","z"])
print(myvar)
```

```python
#DataFrame
data={
    "cal":[420,380,390],
    "duration":[50,40,45]

}
#load data into dataframe object
df=pd.DataFrame(data)
print(df)
```

```python
#locate row
print(df.loc[0])
```

```python
#named index
df=pd.DataFrame(data,index=["day1","day2","day3"])
print(df)
```

```python
#locate named index
print(df.loc["day2"])
```
